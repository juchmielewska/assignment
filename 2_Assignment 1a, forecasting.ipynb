```python
# LIST OF VARIABLES:
#---------------------------------------------------------------------------------------------------------------------------
# Data - date od match.
# Time - time of match.
# Comp - type of competition
# Round - round or phase of competition.
# Day - the day of the week on which the match took place.
# Venue - away (game played at an opponent's ground), home (game played on the home ground).
# Result - match result.
# GF - goals for.
# GA - goals against.
# Opponent - opposing team.
# Poss - possession of the ball.
# Attendance - number of people in the stadium.
# Captain - team captain.
# Formation - how players are positioned on the pitch.
# Referee - main referee of the match.
# Sh - shots total.
# SoT - shots on target.
# Dist - average shot distance.
# Fk - shots from free kicks.
# Pk - penalty kicks made.
# PKatt - penalty kicks attempted.
```

```python
sm["date"] = pd.to_datetime(sm["date"])
sm["gf"] = pd.to_numeric(sm["gf"],errors = 'coerce')
sm["ga"] = pd.to_numeric(sm["gf"],errors = 'coerce')
sm['xg'] = sm['xg'].astype('float64')
sm['xga'] = sm['xga'].astype('float64')
sm['poss'] = sm['poss'].astype('float64')
sm['attendance'] = sm['attendance'].astype('float64')
sm['sh'] = sm['sh'].astype('float64')
sm['sot'] = sm['sot'].astype('float64')
sm['dist'] = sm['dist'].astype('float64')
sm['fk'] = sm['fk'].astype('float64')
sm["pkatt"] = sm["pkatt"].astype('float64')
sm["season"] = sm["season"].astype('int64')
```

```python
sm.dtypes
```

```python
# converting from a string into categories --> and categories into numbers 
sm["venue_code"] = sm["venue"].astype("category").cat.codes

# new column = venue code:
# 0 --> Away
# 1 --> Home
```

```python
# assigning a code to every opponent, new column = opp code
sm["opp_code"] = sm["opponent"].astype("category").cat.codes
```

```python
# converting time column into integer type, new column = hour
sm["hour"] = sm["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
# converting day column into integer type, new column = date
sm["day_code"] = sm["date"].dt.dayofweek

# Monday - 0
# Thuesday - 1
# ...
```

```python
# converting target column into integer type, new column = hour
sm["target"] = (sm["result"] == "W").astype("int")

# Lost [L] --> 0 
# Draw [D] --> 0
# Win [W] --> 1
```

```python
grouped_matches = sm.groupby("team")
group = grouped_matches.get_group("Manchester City").sort_values("date")
group.head()
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = sm.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('team')
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
# matches_rolling.to_csv("matches_rolling.csv")
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-10-01']
    test = data[data["date"] > '2022-10-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
error
```
